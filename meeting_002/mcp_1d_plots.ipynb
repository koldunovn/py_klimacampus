```python
data = ferr.use('air.mon.mean.nc')
```

```python
air = data.gv('air')
```

```python
mcp.shade(air.x, air.y, air[1,:,:]);
```

Calculate global mean for every month

```python
global_mt = air.mean(axis=1).mean(axis=1)
```

Due to bug in numpy.nanmin in numpy 1.6.0 I have to unmask global_mt: 
        
        np.ma.filled(global_mt)

Standard plot with pylab plot

```python
plot(air.t, np.ma.filled(global_mt))
```

Plot with mcp plot1

```python
mcp.plot1(air.t, np.ma.filled(global_mt))
```

```python
mcp.dateplt1(air.t, np.ma.filled(global_mt))
```
