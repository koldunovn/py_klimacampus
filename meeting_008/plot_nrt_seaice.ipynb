```python
import urllib2
import string
import os
import numpy as np
import time
from matplotlib import pylab as plt
%matplotlib inline
```

```python
print("Monitoring run at %s" % time.ctime())
ftpsite='ftp://sidads.colorado.edu/pub/DATASETS/nsidc0081_nrt_nasateam_seaice/north/'
print("Website: "+ftpsite)
u = urllib2.urlopen(ftpsite)
content=u.read(-1)
filename = content.split('\r')[-2].split()[-1]
print("file to be plotted: "+filename)
```

```python
os.system("wget -nc "+ftpsite+filename);
ice = np.fromfile(filename, dtype='uint8')
ice = ice[300:]
ice = ice.reshape(448,304)
print('Today\'s sea ice concentration map:')
ice = ice/250.
ice_masked = np.ma.masked_greater(ice, 1.0)
fig = plt.figure(figsize=(7,10))
plt.imshow(ice_masked);
plt.colorbar(orientation='horizontal', pad=0.05);
```
