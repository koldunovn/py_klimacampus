## a ? b : c 

`a ? b : c` - The expression `a`, before the `?`, is a condition expression, and the ternary operation evaluates the expression `b`, before the `:`, if the condition `a` is `true` or the expression `c`, after the `:`, if it is `false`.

```julia
1 < 2 ? print("yes") : print("no")
```

```julia
danet(x) = (x>0) ? 1 : -1
```

```julia
danet(10)
```

```julia
danet(-10)
```

```julia
using NetCDF
using StatsBase
```

Download file

```julia
#run(`wget ftp://ftp.cdc.noaa.gov/Datasets/ncep.reanalysis/surface/air.sig995.1948.nc`)
```

```julia
ncinfo("air.sig995.1948.nc")
```

```julia
air = ncread ("air.sig995.1948.nc", "air");
```

```julia
air_c = ((air*0.01)+512.81) - 273.15;
```

```julia
size(air_c)
```

```julia
imagesc(squeeze(air_c[:,:,1], 3))
```

```julia
res = Array(Float32,144,73);
```

Native Julia autocorrelation function

```julia
@time(
for i=1:144, j = 1:73
    res[i, j]=autocor(vec(air_c[i,j,:]),[1])[1]
end)
```

Version similar to numpy variant

```julia
@time(
for i=1:144, j = 1:73
    res[i, j]=crosscor(vec(air_c[1,1,2:1464]), vec(air_c[1,2,1:1463]), [0])[1]
end)
```
